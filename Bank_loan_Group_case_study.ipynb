## 1.3 Data Imbalance

For Data imblance check we can take `Target variable`column.

```python
# Now lets check the Target distributions and make countplot to check there is a Data imblance or not and find the ratio of data imblance
ratio = loanprocess2.TARGET.value_counts()
print('No of customer which no difficulty faced represented by 0 is :', ratio[0])
print('No of customer with payment difficulties represented by 1 is :', ratio[1])
print('Ttoal Ratio of Data imbalance is :', round(ratio[0] / ratio[1], 2), ': 1')
ratio.plot(kind='bar', title='Count (TARGET)');
```

#### Find Percentage of Data Imbalance

```python
# now let us check in the number of Percentage Data imblance in Target variable

no_difficulty = len(loanprocess2[loanprocess2["TARGET"]==0]) # no difficulty is represnted by 0
payment_difficulties = len(loanprocess2[loanprocess2["TARGET"]==1]) # client with payment difficulties represented by 1

Percentage_of_no_difficulty = no_difficulty/(no_difficulty + payment_difficulties)
print("Percentage of no difficulty faced is : ",Percentage_of_no_difficulty*100)

Percentage_of_payment_difficulties= payment_difficulties/(no_difficulty+payment_difficulties)
print("percentage of payment difficulties is : ",Percentage_of_payment_difficulties*100)
```
